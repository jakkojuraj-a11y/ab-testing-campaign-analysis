# 📊 Marketing A/B Testing Analysis

**Goal:** Did showing real ads convert more customers than showing a Public Service Announcement (PSA)?

| Column | Meaning |
|---|---|
| `user id` | Unique customer ID |
| `test group` | `ad` = saw real ad · `psa` = saw public service announcement |
| `converted` | `True` if the customer made a purchase |
| `total ads` | How many ads the user saw |
| `most ads day` | Day they saw the most ads |
| `most ads hour` | Hour they saw the most ads |

---

## 1 · Setup
Import all libraries we need in one place.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.proportion import proportions_ztest, confint_proportions_2indep
from statsmodels.stats.power import NormalIndPower
```

Set a consistent visual style for all charts.

```python
plt.rcParams.update({
    "figure.dpi"        : 120,
    "axes.spines.top"   : False,
    "axes.spines.right" : False,
    "axes.titlesize"    : 13,
    "axes.labelsize"    : 11,
})

AD_COLOR  = "#4C72B0"   # blue  — Ad group
PSA_COLOR = "#DD8452"   # orange — PSA group
YES_COLOR = "#2ecc71"   # green — converted
NO_COLOR  = "#e74c3c"   # red   — not converted
```

Load the dataset and drop the auto-generated index column.

```python
df = pd.read_csv("marketing_AB.csv")
df = df.drop(columns=["Unnamed: 0"], errors="ignore")

print(f"Rows   : {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

## 2 · Data Quality

Before any analysis, we check the data is clean and trustworthy.

### 2.1 Column Types & Missing Values
Check data types and see if any values are missing.

```python
quality = pd.DataFrame({
    "Data Type"     : df.dtypes,
    "Missing Count" : df.isnull().sum(),
    "Missing %"     : (df.isnull().sum() / len(df) * 100).round(2),
    "Unique Values" : df.nunique(),
})
quality
```

### 2.2 Duplicate Rows
Duplicate rows can skew results — remove them if found.

```python
dup_rows = df.duplicated().sum()
dup_ids  = df["user id"].duplicated().sum()

print(f"Duplicate rows    : {dup_rows}")
print(f"Duplicate user IDs: {dup_ids}")
```

```python
# Remove duplicates if any exist
if dup_rows > 0:
    df = df.drop_duplicates()
    print(f"Removed {dup_rows} duplicate rows. New shape: {df.shape}")
else:
    print("No duplicates found — dataset is clean.")
```

### 2.3 Outliers in Total Ads
Use the IQR rule: values beyond 1.5×IQR from Q1/Q3 are outliers.

```python
Q1  = df["total ads"].quantile(0.25)
Q3  = df["total ads"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["total ads"] < lower) | (df["total ads"] > upper)]
print(f"Outlier threshold : below {lower:.0f} or above {upper:.0f}")
print(f"Outlier count     : {len(outliers):,} ({len(outliers)/len(df)*100:.1f}%)")
```

### 2.4 Invalid Values Check
Verify hours are 0–23 and days are valid weekday names.

```python
VALID_DAYS    = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
invalid_hours = df[(df["most ads hour"] < 0) | (df["most ads hour"] > 23)]
invalid_days  = df[~df["most ads day"].isin(VALID_DAYS)]
negative_ads  = df[df["total ads"] < 0]

print(f"Invalid hours : {len(invalid_hours)}")
print(f"Invalid days  : {len(invalid_days)}")
print(f"Negative ads  : {len(negative_ads)}")
print()
print("✅ All values are valid." if len(invalid_hours) + len(invalid_days) + len(negative_ads) == 0
      else "⚠️  Some invalid values found — investigate above.")
```

## 5 · A/B Statistical Testing

**Question:** Is the difference in conversion rates real, or just random chance?

- **H₀ (Null):** Ad and PSA groups have the same conversion rate
- **H₁ (Alternative):** Ad group has a higher conversion rate
- **Significance level:** α = 0.05

### 5.1 Prepare the Numbers
Count successes (conversions) and totals per group.

```python
successes = df.groupby("test group")["converted"].sum().values    # [ad, psa]
totals    = df.groupby("test group")["converted"].count().values  # [ad, psa]

print(f"Ad  group : {successes[0]:,} conversions out of {totals[0]:,} users")
print(f"PSA group : {successes[1]:,} conversions out of {totals[1]:,} users")
```

### 5.2 Chi-Square Test
Tests whether the two groups are statistically different.

```python
contingency = pd.crosstab(df["test group"], df["converted"])

chi2, p_chi2, dof, expected = chi2_contingency(contingency)

print(f"Chi² statistic     : {chi2:.4f}")
print(f"Degrees of freedom : {dof}")
print(f"P-value            : {p_chi2:.6f}")
```

### 5.3 Z-Test for Proportions
Another way to test the same question — directly compares two proportions.

```python
z_stat, p_z = proportions_ztest(successes, totals)

print(f"Z-statistic : {z_stat:.4f}")
print(f"P-value     : {p_z:.6f}")
```

### 5.4 Verdict

```python
alpha = 0.05

if p_chi2 < alpha and p_z < alpha:
    print("✅ SIGNIFICANT — The Ad group converts better than PSA (reject H₀)")
else:
    print("❌ NOT SIGNIFICANT — No proven difference between Ad and PSA groups")

print(f"Chi² p-value : {p_chi2:.6f}  ({'< α' if p_chi2 < alpha else '> α'})")
print(f"Z-test p-value: {p_z:.6f}  ({'< α' if p_z < alpha else '> α'})")
```

### 5.5 Business Metrics — How Big is the Difference?

```python
conversion_rates = df.groupby("test group")["converted"].mean() * 100

ad_rate  = conversion_rates["ad"]
psa_rate = conversion_rates["psa"]

abs_lift = ad_rate - psa_rate
rel_lift = (abs_lift / psa_rate) * 100

print(f"Ad  conversion rate : {ad_rate:.2f}%")
print(f"PSA conversion rate : {psa_rate:.2f}%")
print(f"Absolute lift       : +{abs_lift:.2f} percentage points")
print(f"Relative lift       : +{rel_lift:.1f}%")
```

```python
# Effect size (Cohen's h) — how meaningful is the difference?
p1 = ad_rate  / 100
p2 = psa_rate / 100

cohen_h = 2 * np.arcsin(np.sqrt(p1)) - 2 * np.arcsin(np.sqrt(p2))
print(f"Cohen's h (effect size) : {cohen_h:.4f}")
print("Interpretation: <0.2 small · 0.2–0.5 medium · >0.5 large")
```

```python
# 95% Confidence Interval — range where the true difference likely falls
low, high = confint_proportions_2indep(successes[0], totals[0],
                                        successes[1], totals[1])
print(f"95% Confidence Interval : [{low:.4f},  {high:.4f}]")
print("If this range doesn't include 0, the difference is significant.")
```

```python
# Statistical power — how likely we are to detect a real difference
analysis = NormalIndPower()
power = analysis.solve_power(
    effect_size = abs(cohen_h),
    nobs1       = totals[0],
    ratio       = totals[1] / totals[0],
    alpha       = 0.05
)
print(f"Statistical power : {power:.4f}  ({power*100:.1f}%)")
print("Power > 80% means we have enough data to trust the result.")
```

## 6 · Customer Segmentation

Grouping users by how many ads they saw (since revenue data isn't available).

### 6.1 Define Segments
Bottom 10% = Low, Middle 80% = Medium, Top 10% = High.

```python
q10 = df["total ads"].quantile(0.10)
q90 = df["total ads"].quantile(0.90)

print(f"Low exposure  : ≤ {q10:.0f} ads")
print(f"Medium exposure: {q10:.0f} – {q90:.0f} ads")
print(f"High exposure : ≥ {q90:.0f} ads")
```

```python
df["segment"] = pd.cut(
    df["total ads"],
    bins   = [-1, q10, q90, df["total ads"].max()],
    labels = ["Low Exposure", "Medium Exposure", "High Exposure"]
)
```

### 6.2 Segment Summary

```python
seg_summary = df.groupby("segment", observed=True).agg(
    Users       = ("user id",   "count"),
    Conversions = ("converted", "sum"),
    Avg_Ads     = ("total ads", "mean"),
)

seg_summary["Conv Rate (%)"] = (seg_summary["Conversions"] / seg_summary["Users"] * 100).round(2)
seg_summary["Avg_Ads"]       = seg_summary["Avg_Ads"].round(1)
seg_summary
```

```python
seg_summary["Conv Rate (%)"].plot(
    kind="bar",
    color=["#74b9ff", "#0984e3", "#2d3436"],
    edgecolor="white",
    figsize=(6, 4)
)

plt.title("Conversion Rate by Ad Exposure Segment")
plt.ylabel("Conversion Rate (%)")
plt.xlabel("Segment")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
